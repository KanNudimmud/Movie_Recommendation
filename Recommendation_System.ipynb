```python
## Import Libraries
# for data processing
import pandas as pd
import numpy as np
import scipy.stats

# for visualization
import seaborn as sns

# for similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
## Download and Read the Data
# Source: https://grouplens.org/datasets/movielens/ ("ml-latest-small.zip")
```

```python
# Read ratings in the dataset
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
# Get information about data
ratings.info()
```

```python
# Get unique numbers
print('The ratings dataset has', ratings['userId'].nunique(), 'unique users')

print('The ratings dataset has', ratings['movieId'].nunique(), 'unique movies')

print('The ratings dataset has', ratings['rating'].nunique(), 'unique ratings')

print('The unique ratings are', sorted(ratings['rating'].unique()))
```

```python
# Read movies in the dataset
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
# Merge ratings and movies
df = pd.merge(ratings, movies, on='movieId', how='inner')
df.head()
```

```python
## Exploratory Data Analysis (EDA)
# Aggregate by movie
agg_ratings = df.groupby('title').agg(mean_rating=('rating',
                                                     'mean'),
                                      number_of_ratings= ('rating', 'count')).reset_index()

# Keep the movies with over 100 ratings
agg_ratings_GT100 = agg_ratings[agg_ratings['number_of_ratings'] > 100]
agg_ratings_GT100.info()
```

```python
# Check popular movies
agg_ratings_GT100.sort_values(by='number_of_ratings',
                              ascending=False).head()
```

```python
# Use jointplot to see correlation between average rating and # of ratings
sns.jointplot(x='mean_rating', y='number_of_ratings', data = agg_ratings_GT100)
```

```python
# Merge data
df_GT100 = pd.merge(df, agg_ratings_GT100[['title']], on='title',
                    how='inner')
df_GT100.info()
```

```python
# Check numbers after filtering
print('The ratings dataset has', df_GT100['userId'].nunique(), 'unique users')

print('The ratings dataset has', df_GT100['movieId'].nunique(), 'unique movies')

print('The ratings dataset has', df_GT100['rating'].nunique(), 'unique ratings')

print('The unique ratings are', sorted(df_GT100['rating'].unique()))
```

```python
## Create User-Movie Matrix
matrix = df_GT100.pivot_table(index='userId', columns='title',
                              values='rating')
matrix.head()     
```

```python
## Normalization
matrix_norm = matrix.subtract(matrix.mean(axis=1), axis='rows')
matrix_norm.head()
```

```python
## Identify Similar Users
# User similarity matrix using Pearson correlation
user_similarity = matrix_norm.T.corr()
user_similarity.head()
```

```python
# Pick a user ID
picked_userid = 1

# Remove picked used ID from the candidate list
user_similarity.drop(index=picked_userid, inplace=True)

# Let's take a look
user_similarity.head()
```

```python
# Number of similar users
n = 10

# User similarity threshold
user_similarity_threshold = 0.3

# Get top n similar users
similar_users = user_similarity[user_similarity[picked_userid]>user_similarity_threshold][picked_userid].sort_values(ascending=False)[:n]

# Print out top n similar users
print(f'The similar users for user {picked_userid} are', similar_users)
```

```python
## Narrow down item pool
# Movies that the target user has watched
picked_userid_watched = matrix_norm[matrix_norm.index ==
                                    picked_userid].dropna(axis=1,
                                                          how='all')
picked_userid_watched
```

```python
# Movies that similar users watched, remove movies none of the similar users have watched
similar_users_movies = matrix_norm[matrix_norm.index.isin(similar_users.index)].dropna(axis=1,
                                                                                       how='all')
similar_users_movies
```

```python
# Remove the watched movies from the movie list
similar_users_movies.drop(picked_userid_watched.columns,
                          axis=1, inplace=True, errors='ignore')
similar_users_movies
```

```python
## Recommend items
# A dictionary to store item scores
item_score = {}

# Loop over items
for i in similar_users_movies:
  # Get the rating for movie i
  movie_rating = similar_users_movies[i]

  # Create  variables to store the score and # of them
  total = 0
  count = 0

  # Loop over similar users
  for u in similar_users.index:
    # If the movie has rating
    if pd.isna(movie_rating[u]) == False:
      # Score is the sum of user similarity score multiply by the movie rating
      score = similar_users[u] * movie_rating[u]

      # Add the score to total score for the movie so far
      total += score
      count += 1

  # Get the average score for the item i
  item_score[i] = total / count

# Convert dictionary to pandas dataframe
item_score = pd.DataFrame(item_score.items(), columns=['movie',
                                                       'movie_score'])
# Sort the movies by score
ranked_item_score = item_score.sort_values(by='movie_score', ascending=False)

# Select top m movies
m = 10
ranked_item_score.head(m)
```

```python
## Predict Scores
# Average rating for the picked user
avg_rating = matrix[matrix.index == picked_userid].T.mean()[picked_userid]

# Print the average movie rating for user 1
print(f'The average movie rating for user {picked_userid} is {avg_rating:.2f}')
```

```python
# Calcuate the predicted rating
ranked_item_score['predicted_rating'] = ranked_item_score['movie_score'] + avg_rating

# Take a look 
ranked_item_score.head(m)
```
